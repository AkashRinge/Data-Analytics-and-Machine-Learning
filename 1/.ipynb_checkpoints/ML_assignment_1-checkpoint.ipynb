```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import plot_regress_exog
import math
```

```python
os.getcwd()
```

```python
df=pd.read_csv('imports-85.csv').dropna()
df['log_price']=df['price'].apply(lambda x: math.log(x))
df['price_sq']=df['price'].apply(lambda x: x**2)
df.head()
```

```python
#groups = df.groupby('body-style')
#for name, group in groups:
#  print(group)
#  print(name)
```

```python
def plot(x, y, xlabel, ylabel, title):
    groups = df.groupby('body-style')
    for name, group in groups:
        plt.plot(group[x], group[y], marker='o', linestyle='', markersize=7, label=name) 
    plt.legend()
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
plot('horsepower', 'price', 'Horsepower', 'Price', 'Relationship between Price and Horsepower and Body Style')
```

```python
plot('horsepower', 'log_price', 'Horsepower', 'Log Price', 'Relationship between Log Price and Horsepower and Body Style')
```

```python
plot('horsepower', 'price_sq', 'Horsepower', 'Price Squared', 'Relationship between Price squared and Horsepower and Body Style')
```

Plot of Log Price and Horse Power looks linear. Choosing this specification for linear regression, 

```python
y = df['log_price']
X = df['horsepower']
X= sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
residuals = model.resid
fig = plt.figure(figsize=(10, 10))
plot_regress_exog(model, 'horsepower', fig=fig)
plt.show()
```

```python
sm.qqplot(residuals, line='s')
```

```python
plt.figure()
plt.title(" horsepower vs fuel efficiency")
plt.xlabel("Horsepower")
plt.ylabel("Fuel Efficiency")
y=df["city-mpg"]
x=df["horsepower"]
x=sm.add_constant(x)
reg=sm.OLS(y,x).fit()
b,m = reg.params
plt.axline(xy1=(0, b), slope=m, label=f'$y = {m:.1f}x {b:+.1f}$')
plt.scatter(df["horsepower"],df["city-mpg"],color='red')
plt.legend()
```

```python
def ind_plot(x, y, xlabel, ylabel, title):
    groups = df.groupby('body-style')
    fig, axs = plt.subplots(len(groups),1, figsize=(5,5*len(groups)),sharex=True, sharey=True)
    i=0
    for name,group in groups:
        axs[i].plot(group[x], group[y], marker='o', linestyle='', markersize=7, label=name)
        axs[i].set_title(name)
        i+=1
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.suptitle(title)
    plt.subplots_adjust(hspace=0.1)
    plt.show()
```

```python
ind_plot('horsepower', 'price', 'Horsepower', 'Price', 'Relationship between Price and Horsepower and Body Style')
```

```python
df['log_price']=df['price'].apply(lambda x: math.log(x))
ind_plot('horsepower', 'log_price', 'Horsepower', 'Log Price', 'Relationship between Log Price and Horsepower and Body Style')
```

```python
df['price_sq']=df['price'].apply(lambda x: x**2)
ind_plot('horsepower', 'price_sq', 'Horsepower', 'Price Squared', 'Relationship between Price squared and Horsepower and Body Style')
```

```python
StockRetAcct_df = pd.read_csv('StockRetAcct_DT.csv')
```

```python
StockRetAcct_df["excess_ret"] = np.exp(StockRetAcct_df["lnAnnRet"]) - np.exp(StockRetAcct_df["lnRf"])
```

```python
StockRetAcct_df["year"].unique()
```

```python
StockRetAcct_df["year"].min()
StockRetAcct_df["year"].max()
```

```python
StockRetAcct_df['lnIssue_decile'] = np.nan
ret = np.empty([])
EW=pd.DataFrame({'lnIssue_decile':np.arange(1, 11, 1),'MeanExRet':np.empty((10))})

for year in range(1980, 2015):
    dt = StockRetAcct_df[StockRetAcct_df['year'] == year]
    inds = np.where(StockRetAcct_df['year'] == year)
    StockRetAcct_df['lnIssue_decile'].iloc[inds]  = pd.qcut(dt['lnIssue'], 10, labels=np.arange(1, 11, 1)).values.to_list()
    for bin in range(1, 11):
        wt = dt[dt["lnIssue_decile"] == bin].MEwt / dt[dt["lnIssue_decile"] == bin].MEwt.sum()
        ret = np.append(ret, sum(dt[dt.lnIssue_decile == bin].excess_ret * wt))
EW['MeanExRet'] = np.mean(ret[1:].reshape((34, 10)), axis=0)
```
